```python
import sys

sys.path.append("/home/jchen/project-8")
import main as exp
from bovw import BOVW
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_train, map_classes = exp.Dataset(ImageFolder="/home/jchen/project-8/data/train")
data_test, _ = exp.Dataset(ImageFolder="/home/jchen/project-8/data/val", map_classes=map_classes)
test_outputs = {}
```

## Test different amounts of codebook sizes k. What performs best?

```python
codebook_sizes = [100, 200, 400, 800, 1600, 3200, 6400]
```

```python
for codebook_size in codebook_sizes:
    print(f"\n\nTraining with codebook size: {codebook_size}")
    bovw = BOVW(detector_type="SIFT", codebook_size=codebook_size)

    bovw, classifier = exp.train(dataset=data_train, bovw=bovw, map_classes=map_classes, batch_size=100)

    test_outputs[codebook_size] = exp.test(dataset=data_test, bovw=bovw, classifier=classifier, map_classes=map_classes, batch_size=100)
```

```python
# plot average ROC curve for each codebook size
plt.figure(figsize=(10, 8))
for codebook_size, metrics in test_outputs.items():
    all_fpr = np.unique(np.concatenate([metrics["false_positive_rates"][i] for i in range(metrics["num_classes"])]))
    mean_tpr = np.zeros_like(all_fpr)
    
    for i in range(metrics["num_classes"]):
        mean_tpr += np.interp(all_fpr, metrics["false_positive_rates"][i], metrics["true_positive_rates"][i])
    mean_tpr /= metrics["num_classes"]
    plt.plot(all_fpr, mean_tpr, label=f'Codebook {codebook_size} (AUC = {metrics["auc_average"]:.2f})')
plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Average ROC Curve Comparison by Codebook Size')
plt.legend(loc="lower right")
plt.grid()
plt.show()
plt.savefig("roc_curve_codebook_size_comparison.png")
```

```python
# bar plots for each codebook size showing accuracy, precision, recall, f1-score
metrics_names = ["accuracy", "precision", "recall", "f1_score"]
metrics_labels = ["Accuracy", "Precision", "Recall", "F1-Score"]

x = np.arange(len(codebook_sizes))
width = 0.2
fig, ax = plt.subplots(figsize=(12, 8))
for i, (metric_name, metric_label) in enumerate(zip(metrics_names, metrics_labels)):
    values = [test_outputs[codebook_size][metric_name] for codebook_size in codebook_sizes]
    ax.bar(x + i * width, values, width, label=metric_label)
ax.set_xlabel('Codebook Size')
ax.set_ylabel('Scores')
ax.set_title('Classification Metrics by Codebook Size')
ax.set_xticks(x + width * (len(metrics_names) - 1) / 2)
ax.set_xticklabels(codebook_sizes)
ax.legend()
plt.grid(axis='y')
plt.show()
plt.savefig("classification_metrics_codebook_size_comparison.png")
```

## Test different numbers of local features

```python
max_features_list = [50, 100, 200, 400, 800]
codebook_size = 800
detector_types = ['SIFT', 'AKAZE', 'ORB']
local_feature_results = {}

for detector_type in detector_types:
    local_feature_results[detector_type] = {}
    for max_features in max_features_list:
        print(f"\n\nTraining with detector: {detector_type}, max_features: {max_features}")
        bovw = BOVW(detector_type=detector_type, codebook_size=codebook_size, max_features=max_features)

        bovw, classifier = exp.train(dataset=data_train, bovw=bovw, map_classes=map_classes, batch_size=100)

        local_feature_results[detector_type][max_features] = exp.test(
            dataset=data_test,
            bovw=bovw,
            classifier=classifier,
            map_classes=map_classes,
            batch_size=100,
        )

        metrics = local_feature_results[detector_type][max_features]
        print(
            f"{detector_type:5s} | max_features={max_features:4d} | F1={metrics['f1_score']:.3f} | AUC={metrics['auc_average']:.3f}"
        )
```

```python
# Plot F1-score vs number of local features (one line per detector)
plt.figure(figsize=(9, 6))

for detector_type in detector_types:
    f1_scores = [local_feature_results[detector_type][mf]['f1_score'] for mf in max_features_list]
    plt.plot(max_features_list, f1_scores, marker='o', linewidth=2, label=detector_type)

plt.title('F1-score vs Number of Local Features (per Detector)')
plt.xlabel('Number of Local Features (max_features)')
plt.ylabel('F1-score')
plt.grid(True)
plt.xticks(max_features_list)
plt.ylim(0, 1)
plt.legend(title='Detector Type')
plt.show()
```

The experiment gives us this plot. It shows that performance improves when increasing the number of local features, but only up to a certain point.
After a moderate number of features, the improvement becomes small or may even decrease due to noise or redundancy.


## Play with reducing dimensionality
Provem PCA sobre els histogrames BoVW (codebook de 800) per veure l'impacte de reduir la dimensionalitat i com afecta F1/AUC.


```python
# Fit codebook una vegada i precomputa histogrames per a l'experiment de PCA
base_detector = 'SIFT'
codebook_size = 800
max_features = 400

bovw = BOVW(detector_type=base_detector, codebook_size=codebook_size, max_features=max_features)
print('Fitting codebook for dimensionality reduction experiment...')
batch_size = 100
for i in range(0, len(data_train), batch_size):
    batch = data_train[i : i + batch_size]
    descriptors_batch = []
    for image, label, img_path in batch:
        _, descriptors = bovw._extract_features(np.array(image))
        if descriptors is not None and len(descriptors) > 0:
            descriptors_batch.append(descriptors)
    if descriptors_batch:
        bovw._update_fit_codebook(descriptors_batch)

def compute_histograms(bovw, dataset):
    hists, labels = [], []
    for image, label, img_path in dataset:
        _, descriptors = bovw._extract_features(np.array(image))
        if descriptors is not None and len(descriptors) > 0:
            hist = bovw._compute_codebook_descriptor(descriptors, bovw.codebook_algo)
            hists.append(hist)
            labels.append(label)
    return np.array(hists), np.array(labels)

X_train, y_train = compute_histograms(bovw, data_train)
X_test, y_test = compute_histograms(bovw, data_test)
print(f'Train histograms: {X_train.shape}, Test histograms: {X_test.shape}')
```

```python

# Avalua PCA amb diferents dimensions i compara amb no aplicar PCA
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression

pca_components = [32, 64, 128, 256, 512, 700]
dim_results = {}

def fit_and_eval(transform, name):
    clf = LogisticRegression(class_weight='balanced', max_iter=1000)
    model = make_pipeline(transform, clf) if transform is not None else clf
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    dim_results[name] = exp.compute_metrics(y_test, y_pred, map_classes)

# Baseline sense reducció
fit_and_eval(transform=None, name='no_pca')

# PCA amb diferents components (només si inferior a la dimensió original)
for n_comp in pca_components:
    if n_comp >= X_train.shape[1]:
        continue
    pca = PCA(n_components=n_comp, random_state=0)
    fit_and_eval(transform=pca, name=f'pca_{n_comp}')

baseline_f1 = dim_results['no_pca']['f1_score']
print('Dimensionality reduction results:')
for name, metrics in dim_results.items():
    delta = metrics['f1_score'] - baseline_f1
    print(f"{name:8s} | F1={metrics['f1_score']:.3f} | AUC={metrics['auc_average']:.3f} | dF1={delta:+.3f}")
```

```python
# Visualitza impacte en F1/AUC a mesura que baixem la dimensionalitat
labels = []
f1_scores = []
aucs = []
for name in ['no_pca'] + [f'pca_{n}' for n in pca_components if n < X_train.shape[1]]:
    if name not in dim_results:
        continue
    labels.append(name.replace('no_pca', 'No PCA').replace('pca_', 'PCA '))
    f1_scores.append(dim_results[name]['f1_score'])
    aucs.append(dim_results[name]['auc_average'])

x = np.arange(len(labels))
width = 0.35
plt.figure(figsize=(10, 6))
plt.bar(x - width/2, f1_scores, width, label='F1-score')
plt.bar(x + width/2, aucs, width, label='AUC')
plt.xticks(x, labels, rotation=20)
plt.ylabel('Score')
plt.title('Impact of dimensionality reduction (PCA) on performance')
plt.ylim(0, 1)
plt.grid(axis='y')
plt.legend()
plt.tight_layout()
plt.show()
plt.savefig('dimensionality_reduction_performance.png')
```
